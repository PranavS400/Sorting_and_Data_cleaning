```python
df['People']=df['People'].fillna("")
```

```python
def sqft_conversion(x):
    splitting=x.split(' ')
    if len(splitting)==2:
        return splitting[0]
    try:
        return splitting[0]
    except:
        return None
```

```python
def second_name(x):
    splitting=x.split(' ')
    if len(splitting)==2:
        return splitting[1]
    try:
        return splitting[1]
    except:
        return None
```

```python
def Main_name(x):
    splitting=x.split('-')
    if len(splitting)>2:
        return splitting[0]
    try:
        return x
    except:
        return None
```

```python
def Designation(x):
    splitting=x.split('-')
    if len(splitting)>2:
        return splitting[1]
    try:
        return splitting[1]
    except:
        return None
```

```python
df['Name']=df['People'].apply(Main_name)
df1['First Name'] = df['Name'].apply(sqft_conversion)
df1['Last Name'] = df['Name'].apply(second_name)
df1['Designation']=df['People'].apply(Designation)
```
